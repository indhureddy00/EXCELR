```python
import statsmodels.api as sm
import statsmodels.graphics.gofplots as gof
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for generating random data

# Generate some random data to simulate model residuals
# Replace this with your actual model and its fitting process
data = np.random.normal(0, 1, 100)
model = sm.OLS(data, sm.add_constant(np.arange(len(data))))
results = model.fit()

# Now 'results' contains the fitted model, and you can access residuals
qqplot = gof.qqplot(results.resid, line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```
